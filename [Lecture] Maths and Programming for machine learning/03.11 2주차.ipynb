### 1. Get the value of T_mars(화성의 공전주기)

```python
T_synodic = 780  # 회합주기 780days
T_earth = 365.256  # 지구의 공전주기 365days

# 화성의 공전주기를 구하는 식
T_reverse = 1/T_earth - 1/T_synodic   

print(1/T_reverse)  
```

### 2. Derivative Process

- y=3x^2 식에서 x=3에서의 미분

```python
def fl(x):
  fl = 3 *x**2 
  return fl

s = set([1, 0.5, 0.1, 0.01, 0.001, 0.0001, 0.00001])  # 델타x 0에 수렴
x1 =3

for delta_x in s:
  derivative_of_fl_wrt_x = (fl(x1+delta_x) - fl(x1))/delta_x
  print(delta_x, "\t: ", derivative_of_fl_wrt_x)
```

result = 18

- y=4x^3 -3x^2 =2x -7 식에서 x=7에서의 미분

```python
def fl(x):
  fl = 4*x**3 -3*x**2 + 2*x -7 
  return fl

s = set([1, 0.5, 0.1, 0.01, 0.001, 0.0001, 0.00001])
x1 = 7

for delta_x in s:
  derivative_of_fl_wrt_x = (fl(x1+delta_x) - fl(x1))/delta_x
  print(delta_x, "\t: ", derivative_of_fl_wrt_x)
```

result = 548

### 1. Estimate the height of the Pyramid of Khufu.

```python
height_of_pyramid = number_of_foodsteps *length_of_one_footstep * math.tan(radian_angle_of_pyramid) +1  # tan25
print(height_of_pyramid)
```

## 과제

```python
import pandas as pd
from math import pi, sin, cos, tan, sqrt, atan2

import matplotlib.pyplot as plt
```

```python
result = pd.DataFrame(columns=['x_mars', 'y_mars', 'radius_mars', 'theta_mars'])
```

```python
theta1_list = [159, 5, 85, 196, 179, 180, 210, 121, 178]
theta2_list = [115, 323, 41, 153, 136, 136, 167, 76, 135]

phi1_list = [135, 284, 3, 168, 131, 118, 151, 66, 108]
phi2_list = [182, 346, 49, 218, 184, 168, 204, 123, 153]
```

```python
i = 0
for theta_1, theta_2, phi_1, phi_2 in zip(theta1_list, theta2_list, phi1_list, phi2_list):
  theta_1 = theta_1 * (pi/180)
  theta_2 = theta_2 * (pi/180)

  phi_1 = phi_1 * (pi/180)
  phi_2 = phi_2 * (pi/180)

  x_mars = (sin(theta_2) -  sin(theta_1) + tan(phi_1)*cos(theta_1) - tan(phi_2)*cos(theta_2)) / (tan(phi_1) - tan(phi_2))
  y_mars = tan(phi_1)*x_mars + sin(theta_1) - tan(phi_1)*cos(theta_1)
  radius_mars = sqrt(x_mars**2 + y_mars**2)
  theta_mars = atan2(y_mars, x_mars)*180/pi   # atan2, -파이부터 파이까지 가능

  result.loc[i,'x_mars'] = x_mars
  result.loc[i,'y_mars'] = y_mars
  result.loc[i,'radius_mars'] = radius_mars
  result.loc[i,'theta_mars'] = theta_mars
  
  i += 1

  plt.scatter(x_mars, y_mars, color='royalblue')
# plt.show()
plt.savefig('fig.png')
```

```python
result.to_csv('table.csv')
```
