```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D   
import matplotlib.ticker as mticker
from scipy.stats import reciprocal
%matplotlib inline


import pandas as pd
import numpy as np
import random as rn

import matplotlib.pyplot as plt
import plotly_express as px
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from pandas.core.frame import DataFrame

df = pd.read_csv('./flyash.csv')

df.drop('ID', axis=1, inplace=True)
df.drop('LOI', axis=1, inplace=True)
df.drop('Amorphous Alumina', axis=1, inplace=True)
df.drop('Amorphous Silica', axis=1, inplace=True)

df
```

```python
df.rename(columns={'Sum of amorphous aluminosilicates':'amorphous aluminosilicates'}, inplace=True)
df
```

```python
columns = df.columns
```

```python
import itertools
n_column = [0, 1, 2, 3, 4, 5]

ternary_column = list(itertools.combinations(n_column,3))
```

```python
for i in ternary_column:
  up_name = columns[i[0]]
  left_name = columns[i[1]]
  right_name = columns[i[2]]

  sum = df[up_name] + df[left_name] + df[right_name]
  norm_up = df[up_name]*(100/sum) 
  norm_left = df[left_name]*(100/sum)
  norm_right = df[right_name]*(100/sum)    
   
  df_2 = pd.DataFrame({ up_name : norm_up,
                        left_name : norm_left,
                        right_name : norm_right })
  
  df_2['amorphous aluminosilicates'] = df['amorphous aluminosilicates']


  fig = px.scatter_ternary( df_2, 
                            a = df_2.iloc[:,0], b = df_2.iloc[:,1], c = df_2.iloc[:,2],
                            color="amorphous aluminosilicates",
                            size_max=15)
  
  fig.update_layout(
                      # font_family="Courier New",
                      # font_color="#545454",
                      # title_font_family="Courier",
                      # title_font_color="#344050",
                      # legend_title="Classification",
                      # legend_title_font_color="#545454",
                      legend_title_font_size=18,
                      # title_text = '',
                      # title_font_size=24,
                      height = 600, width = 1000)
  
  fig.update_traces(marker={'size': 8})




  fig.show()
```
