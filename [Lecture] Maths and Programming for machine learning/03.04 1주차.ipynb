#### Summation

```python
sum = 0 

for i in range(1,11):  # 혹은 range(11)
  sum = sum +i
print(sum)
```

```python
sum = 0 

for i in range(1,101):   # 혹은 range(101)
  sum = sum +i
print(sum)
```

```python
sum = 0 

s = set([3,5,7])

for i in s :
  sum = sum +i
  print(i)

print(sum)
```

#### Product

```python
sum = 0 


for i in range(1,4):
  sum = sum + (2*i+1)
  print(i)

print(sum)
```

```python
pro = 1 


for i in range(1,101):
  pro = pro * i
  print(pro)

print(pro)
```

#### Artificial Neuron

```python
x1 = 0.1
x2 = 0.3
theta = 0.2

if (x1 + x2) > theta:
  activation = 1

else:
  activation = 0

print(activation)
```
